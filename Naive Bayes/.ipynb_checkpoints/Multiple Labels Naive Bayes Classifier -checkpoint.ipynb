## Naive Bayes with Multiple Labels called  multinomial Naive Bayes classification

**wine dataset** which is a very famous **multi-class classification problem**. 
* "This dataset is the result of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars."                                  [https://archive.ics.uci.edu/ml/datasets/wine]

The Dataset comprises of **13 features**
*  Alcohol
* Malic acid
* Ash
* Alcalinity of ash
* Magnesium
* Total phenols
* Flavanoids
* Nonflavanoid phenols
* Proanthocyanins
* Color intensity
* Hue
* OD280/OD315 of diluted wines
* Proline

* and type of wine cultivar.
 *This data has three type of wine Class_0, Class_1, and Class_3. we will build a model to classify the type of wine.*
